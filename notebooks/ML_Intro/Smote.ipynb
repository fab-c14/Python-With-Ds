```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=1)# nsamples = no of data points, weights = how much is one category
```

```python
# x is independent feature
# y is dependent feature
```

```python
pip install pandas
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
final_df = pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```
