```python
pd.merge(data1,data2,how='cross')
```

```python
data1 = pd.DataFrame({
    'key1':[1,2,4,5,6],
    'key2':[4,5,6,7,8],
    'key3':[3,4,5,6,7]
},index=['a','b','c','d','e'])
```

```python
data2 = pd.DataFrame({
    'key11':[1,2,4,5,6],
    'key22':[4,5,6,7,8],
    'key33':[3,4,5,6,7]
},index=['a','b','h','i','j'])
```

```python
data1.join(data2) # columns overlaps 
```

```python
# join works on indexes but merge on columns
```

```python
data1.join(data2,how='right')
```

```python
data1.join(data2,how='inner')
```

```python
data1.join(data2,how='outer')
```

```python
data1.join(data2,how='cross') # no difference between merge and join but join works on indexes
```

```python
#where there is 10 euro and create catergories 
```

```python
def cat_fare(x):# category of fare
    if x < 10 :
        return 'cheap'
    elif x >=10 and x <20:
        return 'mid'
    else:
        return 'high'
```

```python
df['cat_fare'] = df['Fare'].apply(cat_fare)
```
