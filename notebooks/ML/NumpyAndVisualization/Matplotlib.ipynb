```python
x = np.linspace(0,10,200)
y = np.sin(x)
# control figure size
plt.figure(figsize=(8,4))
plt.plot(x,y,'--b') # --r # red with -- line and # b means blue 
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.grid() # grid will be displayed in graph  background
plt.title("My Data")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

# colours = np.random.rand(50) # colour control
# size control
sizes = 500 * np.random.rand(50)
plt.scatter(x,y,c='b',s=sizes,alpha=.5,marker='x') # alpha repr intensity or opacity
plt.xlabel(" X axis ")
plt.ylabel(" Y axis ")
plt.title("Graph Name")

plt.show()
```

```python
 # 3d graphs
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
plt.figure(figsize=(8,9))
sizes =1000 *  np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection='3d') # i have to do 3d projection
ax.scatter(x,y,z,c='b', s=sizes, alpha=.5)
ax.set_xlabel("x lab")
ax.set_ylabel("y lab")
ax.set_zlabel("z lab")

# ax.show()
```
